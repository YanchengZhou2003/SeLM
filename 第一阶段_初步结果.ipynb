```python
import matplotlib.pyplot as plt

# 数据
block_size = [32, 64, 128, 256]
eu_loss = [0.8, 0.8, 0.8, 0.8]
ct_loss = [4.42, 3.62, 3.53, 2.81]

# 绘图
plt.figure(figsize=(6, 4))
plt.plot(block_size, eu_loss, marker="o", label="eu_loss")
plt.plot(block_size, ct_loss, marker="s", label="ct_loss", color='r', markersize=8)

plt.xlabel("block_size")
plt.ylabel("loss")
plt.title("Validation Loss of the First 32 Tokens vs Block Size")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 数据
block_size_ct = [32, 64, 128, 256]
ct_loss = [4.42, 3.62, 3.53, 2.81]

block_size_eu = [32 * 512 * 3000]  # 单个点
eu_loss = [0.80]

# 绘图
plt.figure(figsize=(6, 4))

# ct 曲线
plt.plot(block_size_ct, ct_loss, marker="s", label="ct_loss", color='r', markersize=8)

# eu 单点
plt.scatter(block_size_eu, eu_loss, marker="o", label="eu_loss", color='b', s=60)

# 对数坐标轴
plt.xscale("log")

plt.xlabel("training tokens (log scale)")
plt.ylabel("loss")
plt.title("Validation Loss of the First 32 Tokens vs Block Size")
plt.legend()
plt.tight_layout()
plt.show()
```
